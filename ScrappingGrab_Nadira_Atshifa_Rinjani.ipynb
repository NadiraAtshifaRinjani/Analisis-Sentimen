```python
all_reviews = []

for score in [1, 2, 3, 4, 5]:
    print(f"ðŸ”Ž Mengambil review bintang {score}...")
    rvs, _ = reviews(
        'com.grabtaxi.passenger',
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=3000
    )
    all_reviews.extend(rvs)

# Masukkan ke DataFrame
df = pd.DataFrame(all_reviews)
```

```python
df
```

```python
df = df[['content','score']]
df
```

```python
def label_sentiment(score):
    if score >= 4:
        return "positive"
    elif score == 3:
        return "neutral"
    else:
        return "negative"

# Terapkan pelabelan
df['sentimen'] = df['score'].apply(label_sentiment)
```

```python
print(df['sentimen'].value_counts())
```

```python
df.to_csv('grabreviews.csv', index=False, encoding='utf-8')
```
